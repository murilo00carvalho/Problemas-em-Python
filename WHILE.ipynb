```python
#Implemente um programa que leia dois valores a e b. O programa deve escrever e somar os valores impares existentes entre a e b.
"""a % 2 != 0: verifica se o número é ímpar.

soma += a: acumula os valores ímpares.

a += 1: avança para o próximo número a cada repetição do while."""


soma_pares = 0
soma_impares = 0

a = int(input("Digite o valor de a: "))
b = int(input("Digite o valor de b: "))

while a <= b:
    if a % 2 == 0:
        print("Valor par:", a)
        soma_pares += a
    else:
        print("Valor ímpar:", a)
        soma_impares += a
    a += 1  # Só uma vez aqui

print("Soma dos pares:", soma_pares)
print("Soma dos ímpares:", soma_impares)
```

```python
#Implemente um programa que leia um valor e verifique se é perfeito.
#Para ser perfeito, ele deve corresponde a soma do seus divisores próprios.
# Exemplo: 6, pois 1 + 2 + 3 = 6

num = int(input("Informe um valor inteiro e positivo: "))
while num<0:
  print("O valor deve ser Positivo.")

  num = int(input("Informe novamente o valor inicial do intervalo:"))

soma = 0
d=1
while d<num/2:
  if num%d ==0:
    soma = soma + d
  d = d + 1

if soma == num: print("Numero perfeito")
else: print("Numero não é perfeito")
```
